### Load Healthcare Data From Excel File To Populate Bronze Level Tables (MySQL)

- bronze_patient
- bronze_patient_encounter


```python
# import libraries
import pandas as pd
import numpy as np
import pymysql
from dotenv import load_dotenv
import os
from sqlalchemy import create_engine
```

```python
# load .env; loads credentials into memeory 
load_dotenv(override=True)
```

```python
host=os.getenv('DB_HOST')
user=os.getenv('DB_USER')
password=os.getenv('DB_PASSWORD')
database=os.getenv('DB_NAME')
```

```python
# load data from Excel sheets
dict_df = pd.read_excel('/Users/wmavery/Documents/Programming/DataAnalyst/UTD/Docs/Capstone/Datasets/diabetic_data_load.xlsx', sheet_name=[0,1])
```

```python
# create patient and patient encounter data frames
patient_df = dict_df[0]
patient_encounter_df = dict_df[1]
```

```python
# connect to MySQL db; load tables using dfs
# extract data from patient_df and load bronze_patient table
connection_string = f"mysql+pymysql://{user}:{password}@{host}/{database}"
try:
    engine = create_engine(connection_string)
   
    with engine.connect() as conn:
       
        patient_df.to_sql(name='bronze_patient', con = conn, chunksize=5000, if_exists = 'replace', method='multi', index = False)
        
        conn.close()

except (Exception) as error:
        error_log   = 'Error writing data to table. Check system file.'                  
        with open('error_log.txt','w') as error:
           error.write(error_log)


```

```python
# perform data check of bronze_patient
connection_string = f"mysql+pymysql://{user}:{password}@{host}/{database}"
try:
    engine = create_engine(connection_string)

    with engine.connect() as conn:
        patient = pd.read_sql('''SELECT * FROM bronze_patient;''', con = conn)

        conn.close()

except (Exception) as error:
        error_log   = 'Error writing data to table. Check system file.'                  
        with open('error_log.txt','w') as error:
           error.write(error_log)


patient.head()
```

```python
# connect to MySQL db; load tables using dfs
# extract data from patient_encounter_df and load bronze_patient_encounter table
connection_string = f"mysql+pymysql://{user}:{password}@{host}/{database}"
try:
    engine = create_engine(connection_string)
   
    with engine.connect() as conn:
       
        patient_encounter_df.to_sql(name='bronze_patient_encounter', con = conn, chunksize=5000, if_exists = 'replace', method='multi', index = False)
        
        conn.close()

except (Exception) as error:
        error_log   = 'Error writing data to table. Check system file.'                  
        with open('error_log.txt','w') as error:
           error.write(error_log)
```
